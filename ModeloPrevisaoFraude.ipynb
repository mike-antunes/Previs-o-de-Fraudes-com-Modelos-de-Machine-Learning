- Agora vamos balancear nossos dados para facilitar os modelos de machine learning que iremos utilizar em breve

- Essa etapa é muito importante para que os dados fiquem balanceados e evite que os modelos de machine learning possam aprender mais com os valores que tiverem mais dados do que outros com menos dados. Assim, teremos modelos mais precisos e eficientes reduzindo grandes chances de víeses

```python
# Seed para reproduzir o mesmo resultado
seed = 100

# Cria o balanceador SMOTE
balanceador = SMOTE(random_state = seed)

# Aplica o balanceador
PREDITORAS_RES, TARGET_RES = balanceador.fit_resample(PREDITORAS, TARGET)
```

```python
# Visualizando o balanceamento da variável TARGET
plt.rcParams["figure.figsize"] = [12.00, 5.00]
plt.rcParams["figure.autolayout"] = True
TARGET_RES.value_counts().plot(kind='bar', title='Possíveis Fraudes x Não Fraudes',color = ['#1F77B4', '#FF7F0E']);
```

- Antes de seguirmos precisamos normalizar as escalas dos nossos dados que estão altamente discrepantes e isso para o nosso modelo de machine learning é de extrema importância para se ter um bom desempenho. **Com o MinMaxScaler nós transformamos as colunas que estão em escalas completamente diferentes uma das outras em uma escala padrão para todas**

```python
# Normalização das Variáveis
Normalizador = MinMaxScaler()
X_treino_normalizados = Normalizador.fit_transform(X_treino)
X_teste_normalizados = Normalizador.transform(X_teste)
```

- Aqui temos uma segunda opção para utilizarmos caso desejamos com a padronização dos dados. **Qual a diferença de normalização e padronização?** A medida de normalização transforma a escala dos dados para que fiquem normalizados entre 0 e 1. Já a padronização utiliza medida de escala diferente, ele transforma a média dos dados em 0 e o desvio padrão sendo 1, e nesse caso teremos valores além do 0 e 1 diferente da normalização. Porém, a regra da padronização é, média dos dados sendo 0 e desvio padrão sendo 1

```python
# Padronizacao das Variáveis
Padronizador = StandardScaler()
X_treino_padronizados = Padronizador.fit_transform(X_treino)
X_teste_padronizados = Padronizador.transform(X_teste)
```
